### Correctly Classified Bike Images

```python
{
    "tags": [
        "hide_input",
    ]
}
im_path = '/nvmescratch/mhull32/unrestricted-adversarial-examples/data_generation/augmented/bicycle'
chdir(im_path)
files = listdir(im_path)

handles = [f'{i}.jpg' for i in correct]
im = [i for i in files if i in handles]
images = [Image.open(f) for f in im]
num_rows = 5
num_cols = 4
fig = plt.figure(dpi=300)
gs = gridspec.GridSpec(num_rows, num_cols, hspace=0.15)

ax = [plt.subplot(gs[i]) for i in range(len(images))]

for i,im in enumerate(images):
    ax[i].imshow(im)
    ax[i].axis('off')
plt.subplots_adjust(wspace=-0.775)
plt.show()
```
