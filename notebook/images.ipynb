```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import numpy as np
import os
from os import listdir
from IPython.display import Image
import pathlib
from pathlib import Path
from mpl_toolkits.axes_grid1 import ImageGrid
import numpy as np
from os import chdir
from PIL import Image
import matplotlib.gridspec as gridspec
```

```python
correct = ['blender_45_white', 'blender_165_white', 'blender_135_white', '6f53a6bd542d0cea', 'blender_120_black', 'blender_135_black', 'blender_0_white', 'blender_150_black', 'blender_0_black', 'blender_165_black', 'blender_45_black', 'blender_15_black', 'blender_150_white', 'blender_120_white',
'blender_30_white', 'blender_15_white', 'blender_105_black', 'blender_30_black', 'blender_105_white']
incorrect = ['blender_60_black', 'blender_75_black', 'blender_60_white', 'blender_90_white', 'blender_75_white', 'blender_90_black']
print(f'correct: {len(correct)}')
print(f'incorrect: {len(incorrect)}')
```

### Correctly Classified Images

```python
im_path = '/nvmescratch/mhull32/unrestricted-adversarial-examples/data_generation/augmented/bicycle'
chdir(im_path)
files = listdir(im_path)

handles = [f'{i}.jpg' for i in correct]
im = [i for i in files if i in handles]
images = [Image.open(f) for f in im]
num_rows = 5
num_cols = 4
fig = plt.figure(dpi=300)
gs = gridspec.GridSpec(num_rows, num_cols, hspace=0.15)

ax = [plt.subplot(gs[i]) for i in range(len(images))]

for i,im in enumerate(images):
    ax[i].imshow(im)
    ax[i].axis('off')
plt.subplots_adjust(wspace=-0.775)
plt.show()
```
